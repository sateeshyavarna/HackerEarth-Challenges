```python
def missing_values_table(df):
	mis_val = df.isnull().sum()
	mis_val_percent = 100 * df.isnull().sum() / len(df)
	mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
	mis_val_table_ren_columns = mis_val_table.rename(columns={0: 'Missing Values', 1: '% of Total Values'})
	mis_val_table_ren_columns = mis_val_table_ren_columns[mis_val_table_ren_columns.iloc[:, 1] != 0].sort_values('% of Total Values', ascending=False).round(2)
	print("Your selected dataframe has " + str(df.shape[1]) + " columns.\n There are " + str(mis_val_table_ren_columns.shape[0]) + " columns that have missing values.")
	return mis_val_table_ren_columns

def removeConstantVariables(data,colsToExclude):
    colsRemoved = list()
    for col in data:
      if col not in colsToExclude and data[col].nunique() == 1:
        colsRemoved.append(col)
    return colsRemoved
  
def accuracy(crossTab):
    tot_Postive = crossTab[0][0] + crossTab[1][1]
    tot_predictions = crossTab[0][0] + crossTab[1][1] + crossTab[1][0] + crossTab[0][1]
    return ((tot_Postive / tot_predictions)) * 100
```

```python
def is_month_end_method(month_end,day):
    if int(month_end) == int(day):
        return 1
    else:
        return 0
    
def is_month_start_method(day):
    if day == 1:
        return 1
    else:
        return 0
    
# quater start - 1,4,7,10
def is_quarter_end_method(month):
    if month%3 == 0:
        return 1
    else:
        return 0
    
# quater end - 3,6,9,12    
def is_quarter_start_method(month):
    if month%3 == 1:
        return 1
    else:
        return 0

def is_year_start_method(month):
    if month == 1:
        return 1
    else:
        return 0
    
def is_week_day_method(dayOfWeek):
#     monday indicates 1 and friday indicates 5
    if int(dayOfWeek) >= 1 and int(dayOfWeek) <= 5:
        return 1
    else:
        return 0
    
def get_date_columns(df,dateFormat,col_name):
  from datetime import datetime
  import calendar
  year = lambda x: datetime.strptime(str(x), dateFormat ).year
  month = lambda x: datetime.strptime(str(x), dateFormat ).month
  day = lambda x: datetime.strptime(str(x), dateFormat ).day
  week_number = lambda x: datetime.strptime(str(x), dateFormat ).strftime('%V')
  day_of_week = lambda x: datetime.strptime(str(x), dateFormat ).strftime('%w')
  day_of_year = lambda x: datetime.strptime(str(x), dateFormat ).strftime('%j')
  seasons = [0,0,1,1,1,2,2,2,3,3,3,0] #dec - feb is winter, then spring, summer, fall etc
  season = lambda x: seasons[(datetime.strptime(str(x), dateFormat ).month-1)]
  month_end = lambda x: calendar.monthrange(datetime.strptime(str(x), dateFormat ).year,datetime.strptime(str(x), dateFormat ).month)[1]

  df[col_name + '_year'] = df[col_name].map(year)
  df[col_name + '_Dayofweek'] = df[col_name].map(day_of_week)
  df[col_name + '_Month'] = df[col_name].map(month)
  df[col_name + '_Day'] = df[col_name].map(day)
  df[col_name + '_Week'] = df[col_name].map(week_number)
  df[col_name + '_Dayofyear'] = df[col_name].map(day_of_year)
  df[col_name + '_season'] = df[col_name].map(season)
  df[col_name + '_month_end'] = df[col_name].map(month_end)
#   df[col_name + '_is_month_end'] = df.apply(lambda x: is_month_end_method(int(x['month_end']),int(x['Day'])), axis=1)
#   df[col_name + '_is_month_start'] = df.apply(lambda x: is_month_start_method(x['Day']), axis=1)
#   df[col_name + '_is_quarter_start'] = df.apply(lambda x: is_quarter_start_method(x['Month']), axis=1)
#   df[col_name + '_is_quarter_end'] = df.apply(lambda x: is_quarter_end_method(x['Month']), axis=1)
#   df[col_name + '_is_year_start'] = df.apply(lambda x: is_year_start_method(x['Month']), axis=1)
#   df[col_name + '_is_week_day'] = df.apply(lambda x: is_week_day_method(x['Dayofweek']), axis=1)
  return df.drop(col_name,axis=1)
```

```python
from sklearn.model_selection import KFold
from lightgbm import LGBMClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import KFold
import gc

clf = LGBMClassifier(
        n_estimators=2000,
        learning_rate=0.1,
        num_leaves=123,
        colsample_bytree=.8,
        subsample=.9,
        max_depth=15,
        reg_alpha=.1,
        reg_lambda=.1,
        min_split_gain=.01,
        min_child_weight=2
    )

folds = KFold(n_splits=5, shuffle=True, random_state=42)

oof_preds = np.zeros(len(X))
sub_preds = np.zeros(len(finalData_test))

features_lgb = list(X.columns)
feature_importance_df_lgb = pd.DataFrame()

for n_fold, (trn_idx, val_idx) in enumerate(folds.split(X)):
    trn_x, trn_y = X[features_lgb].iloc[trn_idx], Y.iloc[trn_idx]
    val_x, val_y = X[features_lgb].iloc[val_idx], Y.iloc[val_idx]
    
    clf.fit(trn_x, trn_y,eval_set= [(trn_x, trn_y), (val_x, val_y)],eval_metric='auc', verbose=250, early_stopping_rounds=150)
    oof_preds[val_idx] = clf.predict(val_x, num_iteration=clf.best_iteration_)
    sub_preds += clf.predict(finalData_test[features_lgb], num_iteration=clf.best_iteration_)/ folds.n_splits
    
#     oof_preds[val_idx] = clf.predict_proba(val_x, num_iteration=clf.best_iteration_)[:, 1]
#     sub_preds += clf.predict_proba(finalData_test[features_lgb], num_iteration=clf.best_iteration_)[:, 1] / folds.n_splits
    
    fold_importance_df_lgb = pd.DataFrame()
    fold_importance_df_lgb["feature"] = features_lgb
    fold_importance_df_lgb["importance"] = clf.booster_.feature_importance(importance_type='gain')
    fold_importance_df_lgb["fold"] = n_fold + 1
    feature_importance_df_lgb = pd.concat([feature_importance_df_lgb, fold_importance_df_lgb], axis=0)
    print(oof_preds[val_idx])
    print("-" * 20 +"LGB Fold:"+str(n_fold)+ "-" * 20)
    print('AUC : %.6f',roc_auc_score(val_y, oof_preds[val_idx]))
    print('####Accuracy:', accuracy_score(val_y, oof_preds[val_idx]))
    print('####Precision:', precision_score(val_y, oof_preds[val_idx]))
    print('####Recall:', recall_score(val_y, oof_preds[val_idx]))
    print('####F1 Score:',f1_score(val_y, oof_preds[val_idx], average='binary'))
    del trn_x, trn_y, val_x, val_y
    gc.collect()
    
pd.DataFrame({'Customer_ID': test_data['Customer_ID'], 'Store_Code': test_data['Store_Code'], 'Prediction' : sub_preds},columns=['Customer_ID', 'Store_Code','Prediction']).to_csv('lgb.csv', index=False)
from google.colab import files
files.download('lgb.csv')
```
